```python
import matplotlib.pyplot as plt
import torch
import numpy as np
import cv2
```

```python
model = torch.hub.load('ultralytics/yolov5',"yolov5s")
```

```python
model
```

```python
# Images
img = 'https://ultralytics.com/images/zidane.jpg'  # or file, Path, PIL, OpenCV, numpy, list

# Inference
results = model(img)

# Results
results.pandas().xyxy[0]["name"].tolist()[0] # or .show(), .save(), .crop(), .pandas(), etc.
```

```python
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
```

```python
names =[]

cap = cv2.VideoCapture(-1)
while cap.isOpened():
    ret,frame = cap.read()
    result = model(frame)
    names.append(result.pandas().xyxy[0]["name"].tolist())
    cv2.imshow("yolo",np.squeeze(result.render()))
    if cv2.waitKey(10)&0XFF == ord("q"):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import uuid
import os
import time
```

```python
IMAGES_PATH = os.path.join("data","image")
labels = ["thanks", "you"]
nbr_img=15
if not os.path.exists(IMAGES_PATH):
    if os.name == 'posix':
        !mkdir -p {IMAGES_PATH}
    if os.name == 'nt':
         !mkdir {IMAGES_PATH}
for label in labels:
    path = os.path.join(IMAGES_PATH, label)
    if not os.path.exists(path):
        !mkdir {path}
```

```python
print(IMAGES_PATH)
```

```python
cap = cv2.VideoCapture(0)
for label in labels:
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    
    # Loop through image range
    for img_num in range(nbr_img):
        print('Collecting images for {}, image number {}'.format(label, img_num))
        
        # Webcam feed
        ret, frame = cap.read()
        
        # Naming out image path
        imgname = os.path.join(IMAGES_PATH, label,label+'.'+str(uuid.uuid1())+'.jpg')
        print(imgname)
        # Writes out image to file q
        cv2.imwrite(imgname, frame)
        
        # Render to the screen
        cv2.imshow('Image Collection', frame)
        
        # 2 second delay between captures
        time.sleep(2)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
